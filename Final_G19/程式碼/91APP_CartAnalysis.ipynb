```python
import pandas as pd
import numpy as np
from datetime import timedelta
from pandas import DataFrame,Series
from factor_analyzer import FactorAnalyzer
```

```python
# one = pd.DataFrame(pd.read_csv("123_new/activityPageView_2018-08-05.csv",low_memory=False))
# M_data = pd.DataFrame(pd.read_csv("MemberData.csv",low_memory=False))
# O_data = pd.DataFrame(pd.read_csv("OrderData.csv",low_memory=False))
OS_data = pd.DataFrame(pd.read_csv("OrderSlaveData.csv",low_memory=False))
Cusseg_data = pd.DataFrame(pd.read_csv("customers-segmentation.csv",low_memory=False)) #客戶芬群資料集
```

```python
#將TradesDateTime轉成Datetime資料型態
OS_data['TradesDateTime'] = pd.to_datetime(OS_data['TradesDateTime'].astype(str).str[0:19], format = "%Y-%m-%d %H:%M:%S") + timedelta(hours = 8 )
#OS_data.head(20)
Cusseg_data.head(20) 
```

```python
# 取出購物資料集中2019年的資料
OS_data[OS_data['TradesDateTime']>=pd.datetime(2019,1,1)]
OS_data[OS_data['TradesDateTime']<=pd.datetime(2020,1,1)]
```

```python
temp_OSdata = OS_data[["TradesGroupCode", "OuterProductSkuCode","Qty",  "Status", "MemberID"]]
Member_data = Cusseg_data[["MemberID", "RFM_Level"]]
# temp_OSdata.head(10)
Member_data.head(10)
# len(Member_data)
```

```python
#這裡是將未完成的訂單從購物資料集中排除
indexStatus = temp_OSdata[ temp_OSdata['Status'] != 'Finish'].index
indexStatus[0:20]
```

```python
temp_OSdata.drop(indexStatus , inplace=True)
temp_OSdata.head(20)
temp_OSdata.shape
```

```python
#擔心資料集中有缺失值，將缺失值自資料集中排除
temp_OSdata.dropna()
temp_OSdata.shape
temp_OSdata = temp_OSdata.reset_index(drop=True)
temp_OSdata.head(20)
```

```python
# 從客戶分群資料集中取出RFM_Level為potential 的顧客Member_ID
indexStatus = Member_data[ Member_data['RFM_Level'] != 'Potential'].index
indexStatus[0:20]
```

```python
Member_data.drop(indexStatus , inplace=True)
Member_data.head(20)
Member_data.shape
```

```python
MID_data = Member_data[["MemberID"]]
MID_data.head(20)
MID_data = MID_data.reset_index(drop=True)
MID_data.head(20)
```

```python
# 然後將上述的potential客群的memberID與處理後的購物紀錄進行merge取出該客群的購物紀錄
testdata_1 = pd.merge(temp_OSdata, MID_data)
testdata_1
```

```python
testBasket_1 = (testdata_1.groupby(["MemberID", "OuterProductSkuCode"])["Qty"].sum().unstack().reset_index().fillna(0).set_index("MemberID"))
testBasket_1
```

```python
#製成potential客群的購買矩陣
# testBasket_1 = testBasket_1.drop(["MemberID"], axis=1)
testBasket_1 = testBasket_1[testBasket_1.columns[testBasket_1.sum()>4]]
testBasket_1
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
testBasket_1 = testBasket_1.applymap(encode_units)
```

```python
testBasket_1.to_csv("testBasket_Potential_5.csv")
```
