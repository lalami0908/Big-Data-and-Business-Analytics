```python
import pandas as pd
# Read the input training data
```

```python
rfm_cai = pd.read_csv("rfm_cai_2.csv")
rfm_cai #算出每個顧客的RFM原始值
```

```python
rfm_cai = rfm_cai.set_index(['MemberID'])
rfm_cai
```

```python
#跑出RFM數值的distribution並觀察
import matplotlib.pyplot as plt
import seaborn as sns
# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1)
sns.distplot(rfm_cai['r'])
# Plot distribution of F
plt.subplot(3, 1, 2) 
sns.distplot(rfm_cai['f'])
# Plot distribution of M
plt.subplot(3, 1, 3)
sns.distplot(rfm_cai['m'])
# Show the plot
plt.show()
```

```python
# --Calculate R and F groups-- 設定RFM的分級方式，並且label上去
# Create labels for Recency and Frequency
r_labels = range(5, 0, -1)
bins=[1,2,3,4,5,15]
f_labels = range(1,6)
# Assign these labels to 4 equal percentile groups 
r_groups = pd.qcut(rfm_cai['r'], q=5, labels=r_labels)
# Assign these labels to 4 equal percentile groups 
f_groups = pd.cut(rfm_cai['f'],bins, labels=f_labels)
# Create new columns R and F 
rfm_cai = rfm_cai.assign(R = r_groups.values, F = f_groups.values)
rfm_cai.head()
```

```python
# Create labels for MonetaryValue 設定RFM的分級方式，並且label上去
m_labels = range(1, 6)
# Assign these labels to three equal percentile groups 
m_groups = pd.qcut(rfm_cai['m'], q=5, labels=m_labels)
# Create new column M
rfm_cai = rfm_cai.assign(M = m_groups.values)
rfm_cai
```

```python
# Concat RFM quartile values to create RFM Segments 設定RFM的分級方式1~5分，並且label上去
def join_rfm(x): return str(x['R']) + str(x['M'])+ str(x['F'])
rfm_cai['RFM_Segment_Concat'] = rfm_cai.apply(join_rfm, axis=1)
rfm = rfm_cai
rfm.head()
```

```python
rfm_count_unique = rfm.groupby('RFM_Segment_Concat')['RFM_Segment_Concat'].nunique()
print(rfm_count_unique.sum())
```

```python
rfm['RFM_Score'] = rfm[['R','F','M']].sum(axis=1)
print(rfm['RFM_Score'].head())
```

```python
#設定我們想找的五個特徵客群的篩選標準，並且label上去
def rfm_level(df):
    if ((df['R'] == 5) and (df['F']==5) and (df['M']==5)):
        return 'Best'
    elif ((df['R'] == 5) and (df['F'] <= 2) and (df['M'] >= 3)):
        return 'New'
    elif ((df['R'] >= 4) and (df['F'] >= 4) and (df['M'] <= 3)):
        return 'Loyal'
    elif  ((df['R'] <= 3) and (df['F'] >= 4) and (df['M'] >= 4)):
        return 'Potential'
    elif ((df['R']==1) and (df['F'] <= 1) and (df['M'] <= 2)):
        return 'Bye'
    else:
        return 'Promising'
# Create a new variable RFM_Level
rfm['RFM_Level'] = rfm.apply(rfm_level, axis=1)
# Print the header with top 5 rows to the console
rfm.head()
rfm.to_csv("customers segmentation.csv")
rfm
```

```python
# Calculate average values for each RFM_Level, and return a size of each segment 
# 計算各個客群的原始RFM值之平均，與我們的給分方式呈現同樣的特徵
rfm_level_agg = rfm.groupby('RFM_Level').agg({
    'r': 'mean',
    'f': 'mean',
    'm': ['mean', 'count']
}).round(1)
# Print the aggregated dataset
print(rfm_level_agg)
rfm_level_agg.to_csv("rfm_level_agg.csv")
```

```python
import squarify
#rfm_level_agg.columns = rfm_level_agg.columns.droplevel()
rfm_level_agg.columns = ['RecencyMean','FrequencyMean','MonetaryMean', 'Count']
#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 9)
squarify.plot(sizes=rfm_level_agg['Count'], 
              label=['Best',
                     'Bye',
                     'Loyal',
                     'New',
                     'Potential',
                     'Promising'], alpha=.6 )
plt.title("RFM Segments",fontsize=18,fontweight="bold")
plt.axis('off')
plt.show()
```
